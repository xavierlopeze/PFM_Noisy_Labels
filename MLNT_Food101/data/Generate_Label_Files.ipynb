```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.txt', header = None, names = ["dir"])
df_test = pd.read_csv('test.txt', header = None, names = ["dir"])
df_train
```

```python
asdf = 'apple_pie/1011328'
```

```python
clases = [x.split('/')[0] for x in df_train.dir]
names = [x.split('/')[1] + '.jpg' for x in df_train.dir]

df_train["class"] = clases
df_train["name"] = names
df_train
```

```python
clases = [x.split('/')[0] for x in df_test.dir]
names = [x.split('/')[1] + '.jpg' for x in df_test.dir]

df_test["class"] = clases
df_test["name"] = names
df_test
```

```python
len(set(df_test["class"]))
```

```python
df_mapping = pd.DataFrame()

df_mapping["class_number"] = [int(x) for x in range(0,101)]
df_mapping["class_name"] = set(df_train["class"])
df_mapping
```

```python
dict_map = df_mapping.set_index("class_name").to_dict()['class_number']
```

```python
class_number = [dict_map[x] for x in df_train["class"]]
df_train["class_number"] = class_number

df_train
```

```python
class_number = [dict_map[x] for x in df_test["class"]]
df_test["class_number"] = class_number

df_test
```

```python
df_full = pd.concat([df_train, df_test])
```

```python
df_full["dir"] = ['/images/' + x + '.jpg' for x in df_full["dir"]]
```

```python
df_full
```

```python
clean_kv = 'clean_label_kv.txt'
df_full[["dir","class_number"]].to_csv(clean_kv, sep=' ', index=False,header = False)
```
